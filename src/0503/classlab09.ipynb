```python
import numpy as np
from numpy.random import normal, uniform
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
#lab04
'''

創建一個波型圖
plt.plot()會依資料的特性畫出波型圖

'''

from random import randrange
samples = [randrange(-32768, 32768) for i in range(int(120))] # 隨機產生-32768~32768 120個樣本 
plt.plot(samples) 
plt.show()
```

```python
#lab05-1
'''
pyplot.hist()會自動統計傳入的data,繪製長條圖

'''

gaussian_numbers = normal(size=1000)
plt.hist(gaussian_numbers)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#lab05-2
'''

透過參數bins可以調整bar的數量
       bins預設為10
       
'''

gaussian_numbers = normal(size=1000)  #normal 產生常態分佈的隨機亂數
plt.hist(gaussian_numbers, bins=20)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#lab06
'''

透過參數histtype可以改變長條圖的外型
透過參數Alpha可以改變bar的透明度

'''
gaussian_numbers = normal(size=1000)                  #normal 產生常態分佈的隨機亂數
uniform_numbers = uniform(low=-3, high=3, size=1000)  #uniform 產生浮點數型態的隨機亂數

plt.hist(gaussian_numbers, bins=20, histtype='stepfilled', normed=True, color='b', label='Gaussian')
plt.hist(uniform_numbers, bins=20, histtype='stepfilled', normed=True, color='r', alpha=0.5, label='Uniform')

plt.title("Gaussian/Uniform Histogram")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
#lab07
'''

透過normpdf()在長條圖上劃一條best fit line(最佳配適線)

'''
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)    #bins每個bar的邊緣值(n+1個) mu母體平均數  sigma標準差
plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)   #背景格線

plt.show()
```

```python
#lab08
'''

在同一張長條圖上可以顯示多個數據集的資料

'''
n_bins = 10
x = np.random.randn(1000, 3)

fig, axes = plt.subplots(nrows=2, ncols=2)  #切成4個子圖
ax0, ax1, ax3, ax2 = axes.flat

colors = ['red', 'yellow', 'green']
ax0.hist(x, n_bins,histtype='bar', color=colors, label=colors)  
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')

ax1.hist(x, n_bins, histtype='bar', stacked=True)
ax1.set_title('stacked bar')

ax2.hist(x, n_bins, histtype='step',  stacked=True, fill=True)
ax2.set_title('stepfilled')

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('different sample sizes')

plt.tight_layout()
plt.show()
```

```python
#lab09-1:
'''

利用model bar()亦可劃出長條圖

'''
N = 5
menMeans = (20, 35, 30, 35, 27)

ind = np.arange(N)  # 定義0~4的陣列，作為x座標
width = 0.35       # 長條圖的寬度

plt.bar(ind, menMeans, width, color='r',yerr=0.000001)
plt.show()
#bar的參數為：(X座標, Y的座標, 寬度, color='顏色') yerr是讓長條圖不要貼到最頂
```

```python
#lab09-2:
'''

利用model bar()劃出顯示多個數據集的長條圖

'''
N = 5
menMeans = (20, 35, 30, 35, 27)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()

rects1 = ax.bar(ind, menMeans, width, color='r', yerr=0.000001)

womenMeans = (25, 32, 34, 20, 25)
rects2 = ax.bar(ind + width, womenMeans, width, color='y', yerr=0.000001)


ax.set_ylabel('Scores')                    #增加y座標的說明
ax.set_title('Scores by group and gender') #增加圖的標題
ax.set_xticks(ind + width)                 #每個x軸說明的間距
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5')) #增加X軸每個長條圖的說明

ax.legend((rects1[0], rects2[0]), ('Men', 'Women')) #標示圖例

#標示Y軸的座標
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),ha='center') #設定每個bar上的數值

autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
#lab10
'''

使用barh()可以劃出水平向的長條圖

'''

N = 5
menMeans = (20, 35, 30, 35, 27)

ind = np.arange(N)  # 定義0~4的陣列，作為x座標
width = 0.35       # 長條圖的寬度

plt.barh(ind, menMeans, width, color='r',xerr=0.000001)#只要使用barh就可展現橫的直方圖
```
