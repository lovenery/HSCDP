```python
#Lab06
'''
from numpy import *

array1 = arange(1,5).reshape(2,2)
array1結果
[1 2]
[3 4]

array2 = arange(2,9,2).reshape(2,2)
array2結果
[2 4]
[6 8]

print('array1為:\n ',array1,'\n\n','array2為:\n ',array2,end = '\n\n')

print('array1與array2的內積為:\n',dot(array1,array2),end = '\n\n')  #內積
[1 2]  X  [2 4] = [1*2+2*6 1*4+2*8]
[3 4]     [6 8]   [3*2+4*6 3*4+4*8]

print('array1與array2的外積為:\n',cross(array1,array2),end = '\n\n') #外積
[2 4] =2x8-4x4   [2 1] =2x3-1x6
[4 8]            [6 3] 

print('array1的轉置矩陣為:\n',array1.T,end = '\n\n')                 #轉置矩陣
print('array2的轉置矩陣為:\n',array2.T)                              #轉置矩陣
'''

#from numpy import *

array1 = arange(1,5).reshape(2,2)
array2 = arange(2,9,2).reshape(2,2)

print('array1為:\n ',array1,'\n\n','array2為:\n ',array2,end = '\n\n')

print('array1與array2的內積為:\n',dot(array1,array2),end = '\n\n')  #內積
print('array1與array2的外積為:\n',cross(array1,array2),end = '\n\n') #外積

print('array1的轉置矩陣為:\n',array1.T,end = '\n\n')                 #轉置矩陣
print('array2的轉置矩陣為:\n',array2.T)                              #轉置矩陣
```
